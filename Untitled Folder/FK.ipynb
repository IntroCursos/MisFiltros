# Modelo de carro va a tener varios estados


## Lo vamos a simplificar a solo uno



## La lectura del gps es ruidoso entonces la salida y es ruidosa

## measurement noise (v)

Vamos a mostrar esta medicion (measurement noise) ruidosa con la letra v. la cual es
una variable aleatoria.



## Ademas hay un ruido en el proceso

El cual tambien es una variable aleatoria, y puede representa los efetos del viento o cambios en la velocidad del carro



Aunque estas variables aleatorias no siguien un patron, usando la teoria de probabilidad podemos decir algo sobre su propiedades promedio.

- **v** for example, es asumida a seguir una distribucion Gausiana, con media **cero y convarianza R **
$$v \sim \mathcal{N}(0,\,R)\,.$$
$$R = \sigma^2_v$$

Esto significa si medimos la posicion del carro digamos 100 veces en la misma localizacion. 

El ruido de esta lectura va a tomar un valor con la mayoria de ellos localizados serca de media cero y localizados fuera de ellos

```python
R=1

v = stats.norm(0,R)
#the rvs (random variable sample) method. It
plt.hist(v.rvs(100),normed=True,bins=30);

#v.std() # 68.3% una desviacion estandar
xmax_min = v.interval(0.99)
x999 =np.linspace(*xmax_min,100)

plt.plot(x999,v.pdf(x999),label='PDF')
plt.fill_between(x999, v.pdf(x999), alpha=0.25)
```

## Process noise 

Similarmente el ruido de proceso w es tambien aleatorio y asume una distribucion Gaussiana con covarianza **Q** 

$$v \sim \mathcal{N}(0,\,Q)\,.$$
$$Q = \sigma^2_w$$

## Si conociermos el modelo del carro



podriamos ejecutar la entrada para estimar la posicion.

Pero esta estimacion tambien no seria perfecta. debido a que estamos estimando a x, cual es incierta dado al ruido del preceso.

Es aqui donde el filtro de kalman viene a jugar, El combina estas dos piezas de informacion para venir con la mejor estiamacion del carro.
En presencia del ruido de proceso y de medicion. 

En el paso de tiempo inicial, k-1, la posicion del carro actual puede ser donde sea alrededor del estimado $\hat{x}_{k-1}$ 





Y esta incertidumbre es describida por su **probability density funcion(pdf).** funcion de probabilidad de densidad,

Lo que el dibujo tambien nos muestra es que el carro va estar ms probable alrededor de la media de la distribucion.



En el siguiente paso, la incertidumbre en la estimacion a incrementado **la cual se muestra con una gran varianza.**



Esto es porque entre el paso de tiempo k-1 y k, el carro puede haber caido sobre un hoyo, o tal vez las ruedas puden haberse deslizado un poco. 

Sin embargo puede haber viajado una distancia diferente que la de nuestro modelo.

Sin embargo otra fuente de informacion en la posicion del carro viene de la medicion. 



Aqui la varianza representa la incertidumbre de la medicion. 
Y de nuevo la posicion del carro puede estar donde sea al rededor de la media.


**Ahora que tenemos la prediccion y la medicion, la pregunta es cual es la mejor estimacin del la posicion del carro?**

Resulta que la manera optima de estimar la posicion del carro es combinando estos dos pedasos de informacion.

Esto se realiza multiplicando las dos funciones de probabilidad juntas. 
El resultado del producto es tambien un funcion Gausiana.


Y la media de esta funcion de densidad de probabilidad nos da el estimado optimo de la posicion del carro. 



```python
# Pequeo ejemplo del producto de funciones de distribucion.
P=3
R=1
v = stats.norm(7,R)
p = stats.norm(0,P)
v_rvs = v.rvs(100)
p_rvs = p.rvs(100)


plt.hist(v_rvs,normed=True,bins=30);
plt.hist(p_rvs,normed=True,bins=30);
#op_rvs = p_rvs*v_rvs
#plt.hist(op_rvs,normed=True,bins=30);

#Grafica azul
xmax_min = v.interval(0.99)
vx999 =np.linspace(*xmax_min,100)
plt.plot(vx999,v.pdf(vx999),label='PDF')
plt.fill_between(vx999, v.pdf(vx999), alpha=0.45);

#Grafica verde
xmax_min = p.interval(0.99)
px999 =np.linspace(*xmax_min,100)
plt.plot(px999,p.pdf(px999),label='PDF')
plt.fill_between(px999, p.pdf(px999), alpha=0.45);
```

Este estimado optimo es encontrado por multiplicar la probabilidad de prediccion y la de medicion, escalando el resultado y calculando la media de la funcion de densidad de probabilidad resultate. 

Computacionalmente la multiplicacion de estas dos funciones de probabilidad de densidad se relacionan a la ecuacion de filtro de kalman discreto.

Es similar a un observador de estado. 



De hecho un filtro de kalman es un timpo de observador de estados, pero es diseado para sistemas estocasticos. 

- Dinamica del carro 
$$x_{k+1} = x_{k} + u$$
$$10 = 0 + u \rightarrow u=10$$

- Modelo del carro que es el predictor de la estimacion del estado 
$$\hat{x}_{k+1} = \hat{x}_{k} + u$$
$$7 = 0 + k*10 \rightarrow k=7/10 $$

```python
N=100

Q = 1 # covarianza de proceso
w = stats.norm(0,Q)
w_rvs = w.rvs(N)

x_k=0

#incertidumbre de la estimacion inicial
inc_x = stats.norm(0,1)
hx_k=inc_x.rvs(N) # =0

u=10
B=7/10

x_k1 = x_k + u + w_rvs
hx_k1 = hx_k + B*u

#Initial state estimate
plt.hist(hx_k,normed=True,color='blue')

#Measurement 
plt.hist(x_k1,normed=True,color='red')

#predicted state estimate
plt.hist(hx_k1,normed=True,color='green');
```

# Aadiendo kalman para el optimal state estimate

Note que estos dos estimaciones de estados son diferentes de uno al otro.


Entonces cambiamos la notacion a:



Tambien llamada estimacion a **priori**, debido que es calculada antes de la
medicion actual sea tomada. 

Podemos rescribir la ecuacion como :



Note que la segunda parte de la ecuacion utiliza la medicion.


Y lo incorpora dentro la prediccion para actualizar la estimacion a priori.

A esto se le conoce como el resultado de la estimacion a **posteriori**



```python
N=100

Q = 1 # covarianza de proceso
w = stats.norm(0,Q)
w_rvs = w.rvs(N)

x_k=0

#incertidumbre de la estimacion inicial
inc_x = stats.norm(0,1)
hx_k=inc_x.rvs(N) # =0


u=10
B=7/10

# Predicted estate estimate or priori estimate
hx_k1 = hx_k + B*u

#Dinamics of the car
x_k1 = x_k + u + w_rvs

#Measurement y= C*x = 1*x 

y = x_k1
C=1
##### Modificar K para valores > 1 o para k<1   #####
K=.5
# Kalman filter
ohx_k1 = hx_k1 + K*(y- C*(hx_k1) )

#Initial state estimate
plt.hist(hx_k,normed=True,color='blue')

#Measurement 
plt.hist(x_k1,normed=True,color='red')

#predicted state estimate
plt.hist(hx_k1,normed=True,color='green');

#Optimal state estimate
plt.hist(ohx_k1,normed=True,color='pink');
```
